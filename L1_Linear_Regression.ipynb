```python
# Plotting our data

data = {'x': [2104, 1416, 1534, 852],
        'y': [400, 232, 315, 178]}

df = pd.DataFrame(data)
slope, intercept, r_value, p_value, std_err = linregress(df['x'], df['y'])

plt.scatter(df['x'], df['y'], label='Training Data')
plt.plot(df['x'], intercept + slope * df['x'], color='red', label='Regression Line')

plt.xlabel('Size (sqft)')
plt.ylabel('Price ($1000s)')
plt.title('Seattle Real Estate Prices')

plt.legend()
plt.show()
```

### The Hypothesis
$ h(x) = \theta_0 + \theta_x $

<div style="text-align: center;">
    <h3>$ h(x) = \theta_0 + \theta_x $</h1>
</div>

$ h(x) = \sum_{j=0}^{2} \theta_j x_j$

where $x_0 = 1$

$\theta = \begin{bmatrix} \theta_0 \\ \theta_1 \\ \theta_2 \end{bmatrix}$ 
$x = \begin{bmatrix} x_0 \\ x_1 \\ x_2\end{bmatrix}$

$ \theta $ = parameters  
$m$ = \# of rows (training examples)  
$x$ = "inputs" / features  
$y$ = "outputs" / target variable  

$(x, y)$ = training examples  
$(x^i, y^i)$ = $i^th$ training example

Choose a $\theta$ such that $h(x) \approx y$ for training example  
we choose a $\theta$ to fit the data we already have


$ \frac{1}{2} \sum_{i=1}^{m} (h_Q (x) - y)^2$  
Want to minimize the difference between prediction and the truth

## Gradient Descent

Start with $\theta$
Say $\theta = \overline{0}$ $\theta = \begin{bmatrix}0 \\ 0 \\ 0\end{bmatrix}$

$\theta_j := \theta_j - \alpha \frac{\partial}{\partial\theta_j} J(\theta)$  
  
$\frac{\partial}{\partial\theta}J(\theta) = \frac{\partial}{\partial\theta_j} \frac{1}{2}(h_\theta(x) - y)^2$  
$= 2 \cdot \frac{1}{2} (h_\theta (x) - y) \cdot \frac{\partial}{\partial\theta}(h_\theta(x)-y)$  
$= (h_\theta (x) - y) \cdot \frac{\partial}{\partial\theta}(\theta_0 x_0 + \theta_1 x_1 + ... + \theta_n x_n)$  

for one training example:  
$= (h(x) - y) \cdot x_j$  

for $m$ examples 
$\theta_j = \theta_j - \alpha \sum_{i=1}^{m}(h_\theta(x^(i)) - y^(i))$  

$\frac{\partial}{\partial \theta_j} J(\theta)$ = $\alpha \sum_{i=1}^{m}(h_\theta(x^(i)) - y^(i))$  

### What is Gradient Descent doing?
  - we start with the naive assumption $J(\theta) = 0$
  - with each piece of the traing set we improve our hypothesis


### Batch Gradient Descent  
  - Traditional Gradient descent is performed with a whole group of training data at once
  - Every piece of training data is analyzed to create the final $\theta_j

### Stochastic Gradient Descent
  - Update your hypothesis as you go
  - Every piece of data in your training set is used one by one to improve our hypothesis
  - Useful for incredibly large data sets (100 mil) where it is infeasible to wait for the whole batch to process
  - Function will never converge on a best fit, algorithm tries to optimize for whatever the last piece of training data

For j=1 to m: {  
$\theta_j = \theta_j - \alpha (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_j^{(i)}$  
}
