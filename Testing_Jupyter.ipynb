```python
import matplotlib.pyplot as plt
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random data for the second scatter plot with a trendline and deviations
num_points = 50
x = np.linspace(0, 10, num_points)
trendline = 2 * x + 1
deviations = np.random.normal(0, 2, num_points)
y = trendline + deviations

# Calculate deviations from the trendline
deviation_values = y - trendline

# Create subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Scatter plot with trendline and deviations
axs[0].scatter(x, y, label='Data with Random Deviations')
axs[0].plot(x, trendline, color='red', label='Trendline', linewidth=2)
axs[0].set_title('Scatter Plot with Trendline and Random Deviations')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].legend()

# Plot deviations
axs[1].scatter(x, deviation_values, color='green', label='Deviations')
axs[1].axhline(0, color='red', linestyle='--', label='Zero Deviation')  # Line at zero for reference
axs[1].set_title('Deviations from Trendline')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Deviation')
axs[1].legend()

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for x and y values
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Define a function for the topography (you can replace this with your own function)
def topography_function(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) * 5  # Example function: a sinusoidal hill

# Calculate the z values using the topography function
z = topography_function(x, y)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the topography surface
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')

# Add colorbar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Topography Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for x and y values
x_vals = np.linspace(-10, 10, 100)
y_vals = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x_vals, y_vals)

# Define symbols for x and y
x_sym, y_sym = sp.symbols('x y')

# Define the original topography function
z_sym = sp.sin(sp.sqrt(x_sym**2 + y_sym**2))

# Calculate the derivative of the topography function with respect to x and y
dz_dx_sym = sp.diff(z_sym, x_sym)
dz_dy_sym = sp.diff(z_sym, y_sym)

# Convert the symbolic expressions to Python functions
dz_dx_func = sp.lambdify((x_sym, y_sym), dz_dx_sym, 'numpy')
dz_dy_func = sp.lambdify((x_sym, y_sym), dz_dy_sym, 'numpy')

# Calculate the z values for the original function and its derivatives
z_vals = np.sin(np.sqrt(x**2 + y**2))
dz_dx_vals = dz_dx_func(x, y)
dz_dy_vals = dz_dy_func(x, y)

# Create subplots with 1 row and 3 columns
fig, axs = plt.subplots(1, 3, figsize=(18, 6), subplot_kw={'projection': '3d'})

# Plot the original topography surface
surface_original = axs[0].plot_surface(x, y, z_vals, cmap='viridis', edgecolor='k')
axs[0].set_title('Original Topography')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].set_zlabel('Z-axis')

# Plot the derivative dz/dx
surface_dx = axs[1].plot_surface(x, y, dz_dx_vals, cmap='viridis', edgecolor='k')
axs[1].set_title('Derivative dz/dx')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].set_zlabel('dz/dx')

# Plot the derivative dz/dy
surface_dy = axs[2].plot_surface(x, y, dz_dy_vals, cmap='viridis', edgecolor='k')
axs[2].set_title('Derivative dz/dy')
axs[2].set_xlabel('X-axis')
axs[2].set_ylabel('Y-axis')
axs[2].set_zlabel('dz/dy')

# Add colorbars
fig.colorbar(surface_original, ax=axs[0], shrink=0.6, aspect=12)
fig.colorbar(surface_dx, ax=axs[1], shrink=0.6, aspect=12)
fig.colorbar(surface_dy, ax=axs[2], shrink=0.6, aspect=12)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```
